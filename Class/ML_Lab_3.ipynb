```python
import pandas as pd

df = pd.read_excel('/content/enjoysport.xlsx')
display(df.head())
```

## Implement find-s


```python
df = df.drop('Example' , axis=1)
x = df.drop('enjoysport' , axis=1)
y = df['enjoysport']
```

```python
def find_s(x, y):
    # Initialize hypothesis with the first positive example
    h = None
    for i in range(len(x)):
        if y[i] == 'Yes':  # Check if it's a positive instance
            if h is None:  # Initialize hypothesis with the first positive example
                h = x.iloc[i].tolist()  # Use the entire row excluding target column
            else:
                for j in range(len(h)):
                    if h[j] != x.iloc[i , j]:  # Compare the features
                        h[j] = '?'  # '?' means it can be any value for this feature
    return h

# Example usage (assuming 'x' is a DataFrame and 'y' is the target series)
hypothesis = find_s(x, y)
print("The maximally specific hypothesis is:")
print(hypothesis)
```

## Test the algorithm


```python
# Create a prediction function
def predict(instance, hypothesis):
    # Compare each feature of the instance with the hypothesis
    for i in range(len(hypothesis)):
        # Skip the comparison if the hypothesis feature is '?'
        if hypothesis[i] != '?' and instance[i] != hypothesis[i]:
            return 'No'  # Instance doesn't match the hypothesis
    return 'Yes'  # Instance matches the hypothesis
```

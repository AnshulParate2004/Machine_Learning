# Survival based on Family size

1. higher the family size higher is the chance of death
2. If a person traveling alone the higher chance of death


````python
df['Family size'] = df['SibSp'] + df['Parch'] + 1 # instead of having siblings , and parent child as two differnt columns we can merge it to make family size
````

````python
# making family type column a categorical column which have value alone , small , big

def defineFamilyType(num):
  if(num == 1):
    return 'alone'
  elif(num > 1 and num <= 4):
    return 'small'
  else:
    return 'big'

df['Family Type'] = df['Family size'].apply(defineFamilyType)

df['Family Type']
````

## Identify and handle outliers

### Subtask:
Detect outliers in relevant numerical features and apply suitable techniques (e.g., capping or transformation) to handle them.


**Reasoning**:
Select relevant numerical columns, calculate IQR, define bounds, identify and cap outliers, and verify the process by checking summary statistics.



````python
numerical_cols = ['Age', 'Individual_Fare']

# removing outliers
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    print(f"Outliers for {col}:")
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    display(outliers)

    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print(f"\nSummary statistics after outlier handling:")
display(df[numerical_cols].describe())
````

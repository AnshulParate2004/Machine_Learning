```python
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Supervised Learning/csv files/Salary_Data.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1].values #X is a feature - what are the conditions
Y = dataset.iloc[:,-1].values #Y is a lable - what to predict
```

# Splitting the dataset into the Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
```

This splits the data over random rows , we can set a random state to make it select the same random set of data each time when same random_state is used

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 1/3,random_state=101) #dividing the whole data into 1:2 as test(33%) set and training set(66%) respectively
```

#Training the Simple Linear Regression model on the Training Set

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression() #Instantiate the LinearRegression Class
```

```python
reg.fit(X_train,Y_train) #.fit is a method in sklearn library which is used to train our model
```

#Predict the Test Result

```python
Y_pred = reg.predict(X_test)
Y_pred #model predicted these values(Y_test) for the X_test data
```

```python
Y_test #While this is our orignal data
```

```python
plt.scatter(X_test,Y_test,color='Red',label='Orignal Test Data')
plt.scatter(X_test,Y_pred,color='Green',label= 'Prediction with Test Data') #as we can notice that the predicted data is like a Straight line
plt.plot(X_train,reg.predict(X_train),color='blue',label='Training Data')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Years of Experience')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

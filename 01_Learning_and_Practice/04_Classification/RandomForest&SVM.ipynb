## Dataset Overview

The dataset contains information about loan applicants and their repayment behavior.  
Each record represents one customerâ€™s loan details, including demographic information, financial indicators, and the final loan repayment outcome.

---

### 1. Columns Description

| Column Name | Description |
|--------------|-------------|
| **customer_id** | Unique identifier for each customer. |

### Removing of Outlier

```python
import pandas as pd

# Select only numerical columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Record original shape
original_shape = df.shape[0]

# Function to remove outliers and count how many are removed per column
outlier_summary = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    before = df.shape[0]
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    after = df.shape[0]

    removed = before - after
    outlier_summary[col] = removed

# Show total rows removed
final_shape = df.shape[0]
total_removed = original_shape - final_shape

print(f"Total rows removed: {total_removed}")
print(f"Remaining rows: {final_shape}\n")

# Show breakdown per column
outlier_df = pd.DataFrame.from_dict(outlier_summary, orient='index', columns=['Rows Removed'])
display(outlier_df)
```

```python
# Calculate and visualize the correlation matrix of numerical features
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# Conclusion: Analyze the correlation matrix to identify relationships between numerical features.
```

### One hot Encoding

```python
# Perform one-hot encoding for categorical variables
df = pd.get_dummies(df, columns=['person_home_ownership', 'loan_intent', 'loan_grade'], drop_first=True)

# Convert the target variable 'Repaid' to numerical (Yes: 1, No: 0)
df['Repaid'] = df['Repaid'].map({'Yes': 1, 'No': 0})

# Display the first 5 rows of the modified DataFrame
display(df.head())

# Conclusion: Categorical variables have been converted to numerical using one-hot encoding, and the target variable has been encoded.
```

```python
# Scale numerical features using StandardScaler
from sklearn.preprocessing import StandardScaler

numerical_cols = ['person_age', 'person_income', 'person_emp_length', 'loan_amnt', 'loan_int_rate', 'loan_percent_income', 'cb_person_cred_hist_length']
scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

# Display the first 5 rows of the scaled DataFrame
display(df.head())

# Conclusion: Numerical features have been scaled to have zero mean and unit variance.
```

# Random Forest

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X = df.drop('Repaid', axis=1)
y = df['Repaid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

# Print the shapes of the resulting sets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

# Conclusion: The data has been successfully split into training and testing sets for model development and evaluation.
```

```python
# Train a Random Forest Classifier model
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Conclusion: A Random Forest model has been trained on the training data.
```

```python
# Train a Support Vector Machine (SVM) model
from sklearn.svm import SVC

svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)

# Conclusion: An SVM model has been trained on the training data.
```

```python
# Evaluate the performance of the trained models
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, confusion_matrix

# Evaluate Random Forest model
rf_predictions = rf_model.predict(X_test)
rf_precision = precision_score(y_test, rf_predictions)
rf_recall = recall_score(y_test, rf_predictions)
rf_f1 = f1_score(y_test, rf_predictions)
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_cm = confusion_matrix(y_test, rf_predictions)

print("Random Forest Model Metrics:")
print(f"Precision: {rf_precision:.4f}")
print(f"Recall: {rf_recall:.4f}")
print(f"F1 Score: {rf_f1:.4f}")
print(f"Accuracy: {rf_accuracy:.4f}")
print("Confusion Matrix:\n", rf_cm)
print("-" * 30)

# Evaluate SVM model
svm_predictions = svm_model.predict(X_test)
svm_precision = precision_score(y_test, svm_predictions)
svm_recall = recall_score(y_test, svm_predictions)
svm_f1 = f1_score(y_test, svm_predictions)
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_cm = confusion_matrix(y_test, svm_predictions)

print("SVM Model Metrics:")
print(f"Precision: {svm_precision:.4f}")
print(f"Recall: {svm_recall:.4f}")
print(f"F1 Score: {svm_f1:.4f}")
print(f"Accuracy: {svm_accuracy:.4f}")
print("Confusion Matrix:\n", svm_cm)


# Conclusion: The Random Forest model outperforms the SVM model based on the evaluation metrics.
# The SVM model shows very poor performance, likely due to the complexity and size of the dataset and the default hyperparameters.
```

# Hyper parameter Random Forest

```python
#  Optimized Random Forest Hyperparameter Tuning using RandomizedSearchCV

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
import pandas as pd

# Define the parameter distributions for random search
param_dist = {
    'n_estimators': randint(100, 300),
    'max_depth': [10, 20, None],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 4)
}

# Create the RandomizedSearchCV object
random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_dist,
    n_iter=20,          # only test 20 random combinations (faster)
    cv=3,               # 3-fold cross-validation
    scoring='f1',       # metric: F1 score
    n_jobs=-1,          # use all CPU cores
    random_state=42,
    verbose=2           # show progress in console
)

# Fit the model on a smaller sample to save time (optional but effective)
X_sample = X_train.sample(frac=0.3, random_state=42)
y_sample = y_train.loc[X_sample.index]

print("Starting RandomizedSearchCV...")
random_search.fit(X_sample, y_sample)

# Display the best parameters and score
print("\nBest parameters found:")
print(random_search.best_params_)

print("\nBest F1 score found:")
print(round(random_search.best_score_, 4))

# Retrieve the best model
best_rf_model = random_search.best_estimator_

# Evaluate on test set
y_pred = best_rf_model.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

print("\nClassification Report on Test Data:")
print(classification_report(y_test, y_pred))
```

**Conclusion:**

The Randomized Search for hyperparameter tuning on the Random Forest model yielded the following best parameters:

*   `max_depth`: None
*   `min_samples_leaf`: 1
*   `min_samples_split`: 2
*   `n_estimators`: 266

The best F1 score achieved during the randomized search on the sample data was approximately 0.7694.

Evaluating the best model on the test data resulted in the following metrics:

*   **Precision (for Repaid=1):** 0.93
*   **Recall (for Repaid=1):** 0.66
*   **F1-Score (for Repaid=1):** 0.77
*   **Accuracy:** 0.92

The confusion matrix shows that the model is quite good at correctly identifying both classes, although there are a notable number of false negatives (loans that were repaid but predicted as not repaid).
